```python
import json
import requests
```

```python
# API - To fetch temperature of a city
city_name = input("Enter city name: ")
api_key = ""  # Enter your API key here remove it after use for security reasons
api_url = f"https://api.openweathermap.org/data/2.5/weather?q={city_name}&appid={api_key}"
get_server_information = requests.get(api_url)
Data = get_server_information.json()
print(Data)
temperature_kelvin = Data['main']['temp']
temperature_celsius = temperature_kelvin - 273.15
print(f"The temperature in {city_name} is {temperature_celsius:.2f}°C")
D = Data['weather'][0]['description']
print(f"The weather description in {city_name} is: {D}")
T = Data['main']
print(f"The temperature details are: {T}")
```
